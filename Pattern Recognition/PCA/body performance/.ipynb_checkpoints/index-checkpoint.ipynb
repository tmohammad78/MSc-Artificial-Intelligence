```python
import numpy as np
import pandas as pd
```

```python

data = pd.read_csv("archive/bodyPerformance.csv")
```

```python
data.head()
```

```python
data.drop(columns=["gender","class"],inplace = True)
```

```python
data.head()
```

```python
def data_normalize(scaled_data):
    for col in scaled_data.columns:
        scaled_data[col] =(scaled_data[col] - scaled_data[col].mean())/scaled_data[col].std()
        return scaled_data
```

```python
scaled_data = data_normalize(data)
```

```python
scaled_data.head()
```

```python
cov_matrix = np.cov(scaled_data,rowvar=False,bias=False)
```

```python
cov_matrix.shape
```

```python
from numpy.linalg import eig
```

```python
w,v = eig(cov_matrix)
```

```python
w
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(["e" + str(i+1) for i in range(len(w))], w)
plt.title("Eigenvalues")
plt.xlabel("Magnitude of the eigenvalue")
```
